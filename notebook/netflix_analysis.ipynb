```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')  # change if needed
df.head()
```

```python
print(df.info())
```

```python
print(df['type'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar', title='Movies vs TV Shows')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', title='Content Release Trend')
plt.show()
```

```python
top_genres = df['listed_in'].value_counts().head(10)
top_genres.plot(kind='bar', title='Top Genres')
plt.show()
```
